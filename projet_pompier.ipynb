```python
# Remplacer les virgules par des points dans la colonne "Longitude_2"
df_merge["Longitude_2"] = df_merge["Longitude_2"].str.replace(',', '.')

# Convertir la colonne "Longitude_2" en nombres flottants
df_merge["Longitude_2"] = df_merge["Longitude_2"].astype(float)

# Afficher la colonne "Longitude_2"
print(df_merge["Longitude_2"])
```

```python
# Remplacer les virgules par des points dans la colonne "Longitude_2"
df_merge["Latitude_2"] = df_merge["Latitude_2"].str.replace(',', '.')

# Convertir la colonne "Longitude_2" en nombres flottants
df_merge["Latitude_2"] = df_merge["Latitude_2"].astype(float)

# Afficher la colonne "Longitude_2"
print(df_merge["Latitude_2"])
```

```python
dt = df_merge[["DateAndTimeMobile", "Latitude_1", "Longitude_1", "Latitude_2", "Longitude_2", "DateAndTimeOfCall", "DateAndTimeMobilised", "mobilisation_time", "ResponseTime"]]
```

```python
dt.head()
```

```python
dt.dropna(axis=0, inplace=True)
```

```python
skim(dt)
```

```python
from geopy.distance import geodesic

# Créer une nouvelle colonne dans df_geo pour stocker les distances
dt['Distance_km'] = None  # Initialiser la colonne à None

# Définir copy_on_write sur True
pd.options.mode.copy_on_write = True

# Boucler à travers chaque ligne du DataFrame
for index, row in dt.iterrows():
    # Coordonnées de la station de pompier
    station_coords = (row['Latitude_2'], row['Longitude_2'])
    
    # Coordonnées du lieu d'incendie
    incident_coords = (row['Latitude_1'], row['Longitude_1'])

    
    # Calculer la distance entre la station de pompier et le lieu d'incendie
    distance = geodesic(station_coords, incident_coords).kilometers
    
    # Stocker la distance dans la colonne 'Distance_km' de la ligne actuelle
    dt.loc[index, 'Distance_km'] = distance

# Afficher le DataFrame mis à jour avec les distances calculées
```

## Modelisation: Machine Learning

### - Model 1 avec y1

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import SGDRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y1_encoder, test_size=0.2, random_state=0)
```
